# 2. Membuat statistik deskriptif dari data (setiap atribut yang bertipe numerik)
**2.1 Mean**



Mean merupakan salah satu ukuran pemusatan data, cara menghitung nya yaitu semua nilai dijumlahkan kemudian dibagi dengan total nilai yang di hitung

Rumus mean

$$
      mean = \frac{1}{n} \sum_{i=i}^{n} x_{i}
$$

**2.2 Median**

Median juga salah satu ukuran pemusatan data. cara menghitungnya adalah semua data diurutkan kemudian kita ambil nilai tengah nya


$$
\mathcal {median} = tb + (\frac {\frac {n} 2 - F } {Fm} ) p
$$

Penjelasan :
  - Tb = tepi bawah dari kelas n/2
  - F = Frekuensi kumulatif sebelum kelas median
  - Fm = frekuensi kelas median
  - p = interval

**2.3 Modus**

Modus adalah nilai yang sering muncul


**2.4 Perhitungan**


```python
#Mean
dataset["Length"].mean()
```

```python
#median
dataset["Length"].median()
```

```python
#Modus
dataset["Length"].mode()
```

# 3. Mendeskripsikan kemiringan / skewned dari data setiap variable (bertipe numerik)

Skewness adalah ukuran ketidaksimetrisan dalam distribusi nilai. Skewness dapat bernilai positif, negatif, dan nol.

Skewness yang bernilai positif berarti ekor distribusi berada di sebelah kanan nilai terbanyak. Berarti, sebagian besar distribusi berada di nilai rendah. Skewness yang bernilai negatif berarti ekor distribusi berada di sebelah kiri, menunjukkan bahwa sebagian besar nilai berada di sisi kanan kurva. Sementara skewness bernilai nol berarti nilai terdistribusi secara simetris, dengan jarak antara ekor distribusi sebelah kanan dan kiri sama besar.
$$
skewness =\frac{\sum_{i=1}^{n}\left(x_{i}-\bar{x}\right)^{3}}{(n-1) s^{3}}
$$

Penjelasan 

- $\bar{x}$ adalah sampel mean
- s adalah standar deviasi
- $x_{i}$ adalah sampel


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
sns.distplot(dataset["Length"])
plt.show()
```
