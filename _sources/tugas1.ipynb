**1. Membaca data dari url**

```python
import pandas as pd
```

```python
#Data Gempa
url = "https://vincentarelbundock.github.io/Rdatasets/csv/openintro/earthquakes.csv"
dataset = pd.read_csv(url)
```

```python
#Menampilkan informasi tipe data di semua kolom
dataset.info()
```

```python
# Menampilkan 5 baris teratas
dataset.head()
```

 **2. Membuat statistik deskriptif dari data (setiap atribut yang bertipe numerik)

```python
#Mean
dataset["year"].mean()
```

```python
#median
dataset["year"].median()
```

```python
#Modus
dataset["year"].mode()
```
